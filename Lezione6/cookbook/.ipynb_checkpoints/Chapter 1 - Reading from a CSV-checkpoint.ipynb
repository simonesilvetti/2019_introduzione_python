## 1.0 pandas Series

```python
import pandas as pd 

# creare serie da liste
print(pd.Series([1,2,3]))
print(pd.Series([1,"ddd",3]))
print(pd.Series(["mele","pere","susine"]))
```

```python
#creare serie da dizionario
a = pd.Series({"a":1,"b":2})
print(a)
```

```python
# accedere agli elementi
print(a.loc["a"]) # tramite l'indice 
a.iloc[0]         # tramite la riga
```

```python
# accedere in modo veloce
a["a"]  # equivale a loc
a[0]    # equivale a .iloc
```

```python
# cambiare indice di una serie
print(pd.Series([1,2,3], index = ["a","b","c"]))
```

```python
# unire due serie
s1 = pd.Series({'falcon': 330.0, 'eagle': 160.0})
s2 = pd.Series({'falcon': 345.0, 'eagle': 200.0, 'duck': 30.0})
s1.combine(s2, min)
```

```python
s1.combine(s2, min, fill_value=0)
```

```python
# facciamo la somma
s1.combine(s2, lambda x,y: x+y, fill_value=0)
```

```python
s1 = pd.Series({'falcon': 330.0, 'eagle': 160.0})
s2 = pd.Series({'falcon': 345.0, 'eagle': 200.0, 'duck': 30.0})
s1.combine(s2, max)
```

# 1.1 Reading data from a csv file

You can read data from a CSV file using the `read_csv` function. By default, it assumes that the fields are comma-separated.

We're going to be looking some cyclist data from Montréal. Here's the [original page](http://donnees.ville.montreal.qc.ca/dataset/velos-comptage) (in French), but it's already included in this repository. We're using the data from 2012.

This dataset is a list of how many people were on 7 different bike paths in Montreal, each day.

```python
broken_df = pd.read_csv('../data/bikes.csv') #problema di encoding
```

```python
broken_df = pd.read_csv('../data/bikes.csv',encoding = 'ansi') #problema di encoding
```

```python
# Look at the first 5 rows
broken_df.head() #equivale a broken_df[:5]
```

You'll notice that this is totally broken! `read_csv` has a bunch of options that will let us fix that, though. Here we'll

* change the column separator to a `;`
* Set the encoding to `'latin1'` (the default is `'utf8'`)
* Parse the dates in the 'Date' column
* Tell it that our dates have the day first instead of the month first
* Set the index to be the 'Date' column

```python
fixed_df = pd.read_csv('../data/bikes.csv', sep=';', encoding='ansi', parse_dates=['Date'], dayfirst=True, index_col='Date')

fixed_df.head()
```

```python
len(fixed_df) # numero di righe
```

```python
fixed_df.info() #ottengo informazioni sul dataframe!
```

# 1.3 Plotting a column

Just add `.plot()` to the end! How could it be easier? =)
